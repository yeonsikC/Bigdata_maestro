```python
import csv
import numpy as np
```

```python
# 미국 대통령 평균신장은 얼마일까?
```

```python
names = []
heights = []
index = 0
with open( 'president_heights.csv', 'r' ) as f:
    reader = csv.reader( f )
    for row in reader:
        if index != 0:
            names.append( row[ 1 ] )
            heights.append( int( row[ 2 ] ) )
        index += 1

print( names )
print()
print( heights )
```

```python
name_array = np.array( names )
height_array = np.array( heights )
```

```python
print( 'Mean height              : {0:.3f}\n'.format( height_array.mean() ) )
print( 'Minimum height           : {0}'.format( height_array.min() ) )
print( 'Minimum index            : {0}'.format( height_array.argmin() ) )
print( 'Minimum height president : {0}\n'.format( name_array[ height_array.argmin() ] ) )
print( 'Maximum height           : {0}'.format( height_array.max() ) )
print( 'Maximum index            : {0}'.format( height_array.argmax() ) )
print( 'Maximum height president : {0}\n'.format( name_array[ height_array.argmax() ] ) )

print( 'Standard deviation       : {0:.5f}'.format( height_array.std() ) )
print( '25th percentile          : {0}'.format( np.percentile( height_array, 25 ) ) )
print( 'Median                   : {0}'.format( np.median( height_array ) ) ) 
print( '75th percentile          : {0}'.format( np.percentile( height_array, 75 ) ) )
```

```python
# 미국 시에틀 2014년 비온날 세기
```

```python
dates = []
PRCPs = []
index = 0
with open( 'Seattle2014.csv', 'r' ) as f:
    reader = csv.reader( f )
    for row in reader:
        if index != 0:
            dates.append( row[ 2 ] )
            PRCPs.append( float( row[ 3 ] ) )
        index += 1

print( dates )
print()
print( PRCPs )
```

```python
date_array = np.array( dates )
rainfall_array = np.array( PRCPs )
```

```python
rainfall_array.shape
```

```python
inches = rainfall_array / 254
inches.shape
```

```python
# 비가 온 모든 날에 대한 마스크 생성
rainy = ( inches > 0 )
rainy_date = date_array[ rainy ]
print( rainy_date )
print()

# 여름에 해당하는 날에 대한 마스크 생성( 6월 21일은 172번재 날 )
summer = ( np.arange( 365 ) - 172 < 90 ) & ( np.arange( 365 ) - 172 > 0 )
summer_date = date_array[ summer ]
print( summer_date )
print()
```

```python
# 강수량
print( 'Mean precip on rainy days in 2014 ( inches )        : {0}'.format( np.mean( inches[ rainy ] ) ) )
print( 'Mean precip on summer days in 2014 ( inches )       : {0}\n'.format( np.mean( inches[ summer ] ) ) )

# 여름 평균 강수량
print( 'Maximum precip on summer days in 2014 ( inches )    : {0} - {1}'.format( summer_date[ np.argmax( inches[ summer ] ) ],
                                                                                 np.max( inches[ summer ] ) ) )
# 여름 이외 평균 강수량
print( 'Median precip on non-summer days in 2014 ( inches ) : {0}'.format( np.median( inches[ rainy & ~summer ] ) ) )
```
