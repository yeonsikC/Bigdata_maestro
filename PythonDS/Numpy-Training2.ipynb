```python
import csv
import numpy as np
```

```python
# 전국 주요 상권 정보
```

```python
dates = []
areas = []
cities = []
male_10 = []
male_20 = []
male_30 = []
male_40 = []
male_50 = []
female_10 = []
female_20 = []
female_30 = []
female_40 = []
female_50 = []

index = 0
with open( 'korea_floating_population_data.csv', 'r' ) as f:
    reader = csv.reader( f )
    for row in reader:
        if index != 0:
            dates.append( row[ 3 ] )
            areas.append( row[ 7 ]  )
            cities.append( row[ 7 ].split()[ 0 ] )
            male_10.append( int( row[ 9 ] ) )
            male_20.append( int( row[ 10 ] ) )
            male_30.append( int( row[ 11 ] ) )
            male_40.append( int( row[ 12 ] ) )
            male_50.append( int( row[ 13 ] ) )
            female_10.append( int( row[ 14 ] ) )
            female_20.append( int( row[ 15 ] ) )
            female_30.append( int( row[ 16 ] ) )
            female_40.append( int( row[ 17 ] ) )
            female_50.append( int( row[ 18 ] ) )
        index += 1

print( dates )
print()
print( areas )
print()
print( cities )
print()
print( male_10 )
print()
print( female_10 )
```

```python
date_array = np.array( dates )
area_array = np.array( areas )
city_array = np.array( cities )
male_array = np.array( [ male_10, male_20, male_30, male_40, male_50 ] )
male_array = male_array.T
female_array = np.array( [ female_10, female_20, female_30, female_40, female_50 ] )
female_array = female_array.T
```

```python
male_total = np.sum( male_array[ city_mask ] )
female_total = np.sum( female_array[ city_mask ] )

total = male_total + female_total
print( '대전광역시 주요 상권 방문자 수                   : {0}\n'.format( total ) )

male_age_total = np.sum( male_array[ city_mask ], axis = 0 )
print( '대전광역시 주요 상권 남자 총 방문자 수           : {0}'.format( male_total ) )
print( '대전광역시 주요 상권 남자 방문자 비율            : {0:.2f}%\n'.format( male_total / total * 100 ) )
print( '대전광역시 주요 상권 남자 10대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( male_age_total[ 0 ], 
                                                                                   male_age_total[ 0 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 남자 20대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( male_age_total[ 1 ],
                                                                                   male_age_total[ 1 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 남자 30대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( male_age_total[ 2 ],
                                                                                   male_age_total[ 2 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 남자 40대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( male_age_total[ 3 ], 
                                                                                   male_age_total[ 3 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 남자 50대 방문자 수( 비율 ) : {0}( {1:.2f} )\n'.format( male_age_total[ 4 ], 
                                                                                     male_age_total[ 4 ] / male_total * 100 ) )


female_age_total = np.sum( female_array[ city_mask ], axis = 0 )
print( '대전광역시 주요 상권 여자 총 방문자 수           : {0}'.format( female_total ) )
print( '대전광역시 주요 상권 여자 방문자 비율            : {0:.2f}%\n'.format( female_total / total * 100 ) )
print( '대전광역시 주요 상권 여자 10대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( female_age_total[ 0 ], 
                                                                                   female_age_total[ 0 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 여자 20대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( female_age_total[ 1 ],
                                                                                   female_age_total[ 1 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 여자 30대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( female_age_total[ 2 ],
                                                                                   female_age_total[ 2 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 여자 40대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( female_age_total[ 3 ], 
                                                                                   female_age_total[ 3 ] / male_total * 100 ) )
print( '대전광역시 주요 상권 여자 50대 방문자 수( 비율 ) : {0}( {1:.2f} )'.format( female_age_total[ 4 ], 
                                                                                   female_age_total[ 4 ] / male_total * 100 ) )
```
