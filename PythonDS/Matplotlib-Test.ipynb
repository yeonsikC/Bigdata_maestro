```python
#원하는 지점에 설명을 달아줌.
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('spx.csv', index_col = 0, parse_dates = True)
spx = data['SPX']

spx.plot(ax = ax, style = 'k-')

#원하는 지점에 설명을 달아줌.
crisis_data = {
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
}

# arrowprops : 화살표
for date, label in crisis_data:
    ax.annotate(label, xy = (date, spx.asof(date) + 75),
                xytext = (date, spx.asof(date) + 225),
                arrowprops = dict(facecolor = 'black', headwidth = 4, width= 2,
                                  headlength = 4),
                horizontalalignment = 'left', verticalalignment = 'top')
    
# x, y좌표값 범위
ax.set_xlim([ '1/1/2007', '1/1/2011'])
ax.set_ylim([ 600, 1800])

# title
ax.set_title('Important dates in the 2008 - 2009 financial crisis')

    
```

```python
#그래프에 도형 그리기
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#시작점 : (x=0.2, y=0.75), x넓이 = 0.4, y높이 = 0.15, color = 검은색, 투명도 = 0.3
rect = plt.Rectangle( (0.2, 0.75), 0.4, 0.15, color = 'k', alpha = 0.3)
circle = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha = 0.3)
polygon = plt.Polygon( [[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                     color = 'g', alpha = 0.5)

ax.add_patch(rect)
ax.add_patch(circle)
ax.add_patch(polygon)
```

```python
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]
party_pcts = party_counts.div(party_counts.sum(1), axis = 0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
#히스토그램 밀도그래프 혼합
comp1 = np.random.normal(0, 1, size = 200)
comp2 = np.random.normal(10, 2, size = 200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins = 100, color = 'k')
```

```python
macro = pd.read_csv('macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data = trans_data)
plt.title('Change in log %s versus log %s' % ( 'm1', 'unemp' ))
```

```python
#중요 #산포도와 히스토그램 같이 나타냄
sns.pairplot(trans_data, diag_kind = 'kde', plot_kws = {'alpha':0.2})
```
