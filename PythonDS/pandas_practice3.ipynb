# 역대 미국대통령들의 키에 대한 분석
## 1. 역대 미국대통령들의 키에 대한 기초통계 분석
## 2. 역대 미국대통령들 중 가장 큰 키를 가진 대통령 찾기
## 3. 역대 미국대통령들 중 가장 큰 키를 가진 대통령 찾기

```python
df = pd.read_csv("president_heights.csv")
df.head()
```

```python
print("# 역대 미국 대통령들의 키")
df['height(cm)'].describe()
```

```python
print("# 역대 미국 대통령들 중 가장 작은 키를 가진 대통령")
df[df["height(cm)"]==min(df["height(cm)"])]
```

```python
print("# 역대 미국 대통령들 중 가장 큰 키를 가진 대통령")
df[df["height(cm)"]==max(df["height(cm)"])]
```

# 한국 상권(유통) 분석
## 나의 동네의 주 행인의 성별과 나이대를 알아보자.

```python
df = pd.read_csv("korea_floating_population_data.csv", encoding="CP949")
df.head()
```

```python
my = df[df.행정구역명 == "대전광역시 서구 괴정동"]
my
```

```python
gender_age = []
percent = []
chk = 0
for i in range(0,len(my)):
    for j in range(9,18):
        if my.iloc[i,j] == my.iloc[i,9:18].max():
            chk = j
            break
    gender_age.append(my.columns[chk])
    percent.append(round(my.iloc[i,chk]/sum(my.iloc[i,9:18])*100))
print(gender_age)
print()
print(percent)
```

```python
my_df = pd.DataFrame({"행정구역명":my.행정구역명, "조사일자":my.조사일자,
                      "시간대":my.시간대, "주구분":my.주구분, "날씨":my.날씨,
                      "타겟":gender_age, "비율(%)":percent})
print("# 조사일자, 조사 시간대 별, 가장 많이 움직이는 행인 성별 및 나이대")
my_df
```
